### 1. Best Selling Products


```sql
select p.product_name,p.brand, round(sum(f.total_amount),1)as total_revenue
from learning_pradeep.pipeline.silver_fact_sales f
left join learning_pradeep.pipeline.silver_dim_products p on f.product_sk = p.product_sk
group by p.product_name,p.brand
order by total_revenue desc limit 10
```

### 2. Category Performance


```sql
select p.category, round(sum(f.total_amount),1)as total_revenue

from learning_pradeep.pipeline.silver_fact_sales f
left join learning_pradeep.pipeline.silver_dim_products p on f.product_sk = p.product_sk
group by p.category
order by total_revenue desc limit 10
```

### 3. Brand Contribution


```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_products p  on f.product_sk = p.product_sk

select p.brand , round(sum(f.total_amount),1) as total_revenue

group by p.brand
order by total_revenue desc limit 10
```

### 4. High vs Low Value Products


```sql
select sales_id,product_sk,total_amount from learning_pradeep.pipeline.silver_fact_sales limit 20
```

```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_products p on f.product_sk = p.product_sk

select 
  case 
  when f.total_amount < 2000 then 'affordable'
  when f.total_amount > 2000 then 'medium'
  else 'niche' end as product_type ,

  round(avg(f.total_amount),1) as total_revenue,
  count (*) as total_sales

group by product_type 

order by total_revenue desc   
```
