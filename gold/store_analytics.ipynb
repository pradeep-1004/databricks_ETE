# STORE ANALYTICS

```sql
select * from learning_pradeep.pipeline.silver_dim_stores limit 10
```

### 1.Total Revenue by Store

```sql
select s.store_name , round(sum(f.total_amount),2) as total_revenue 
from learning_pradeep.pipeline.silver_fact_sales f
left join learning_pradeep.pipeline.silver_dim_stores s on f.store_sk = s.store_sk
group by s.store_name 
order by total_revenue desc 
```

### 2. Revenue by City


```sql
select s.store_location , round(sum(f.total_amount),2) as total_revenue
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_stores s on f.store_sk = s.store_sk
group by s.store_location
order by total_revenue desc
```

### 3. Top 10 Stores by Revenue


```sql
select s.store_name , round(sum(f.total_amount),2) as total_revenue 
from learning_pradeep.pipeline.silver_fact_sales f
left join learning_pradeep.pipeline.silver_dim_stores s on f.store_sk = s.store_sk
group by s.store_name 
order by total_revenue desc limit 10
```

### 4. Underperforming Stores


```sql
select s.store_name, round(sum(f.total_amount),2) as revenue
from learning_pradeep.pipeline.silver_fact_sales f
left join learning_pradeep.pipeline.silver_dim_stores s on f.store_sk = s.store_sk
group by s.store_name
having sum(f.total_amount) < 0.25 * (
    select sum(total_amount)
    from learning_pradeep.pipeline.silver_fact_sales
)
order by revenue asc
```

### 5. Campaign Impact by Store


```sql
select s.store_name,c.campaign_name,round(sum(f.total_amount),1) as total_revenue

from learning_pradeep.pipeline.silver_fact_sales f
left join learning_pradeep.pipeline.silver_dim_stores s on f.store_sk = s.store_sk
left join learning_pradeep.pipeline.silver_dim_campaigns c on f.campaign_sk = c.campaign_sk

group by s.store_name,c.campaign_name 
order by total_revenue desc limit 10
```
