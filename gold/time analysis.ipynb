### Yearly Trend

```sql
select * from learning_pradeep.pipeline.bronze_fact_sales_normalized 
order by sales_date desc limit 10
```

```sql
select * from learning_pradeep.pipeline.silver_dim_dates 
order by date_sk desc limit 10
```

```sql
from learning_pradeep.pipeline.bronze_fact_sales_normalized f 
left join learning_pradeep.pipeline.silver_dim_dates d on f.sales_date = d.full_date  

select d.year , sum(f.total_amount) as revenue
group by d.year
order by revenue limit 10
```

### Monthly Trend

```sql
from learning_pradeep.pipeline.bronze_fact_sales_normalized f 
left join learning_pradeep.pipeline.silver_dim_dates d on f.sales_date = d.full_date 

select d.month ,d.year, sum(f.total_amount) as revenue
group by d.month,d.year
order by month limit 10
```

### Quarter Trend

```sql
from learning_pradeep.pipeline.bronze_fact_sales_normalized f 
left join learning_pradeep.pipeline.silver_dim_dates d on f.sales_date = d.full_date 

select d.quarter , sum(f.total_amount) as revenue
group by d.quarter
order by revenue limit 10
```

### Weekday vs Weekend

```sql
from learning_pradeep.pipeline.bronze_fact_sales_normalized f 
left join learning_pradeep.pipeline.silver_dim_dates d on f.sales_date = d.full_date 

select d.weekday , round(sum(f.total_amount) ,1 ) as weekday_rev 
group by d.weekday
order by weekday asc limit 10

```
