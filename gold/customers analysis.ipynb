### Top Customers


```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_customers c on f.customer_sk = c.customer_sk

select c.first_name , round(sum(f.total_amount),1) as total_revenue
group by c.first_name 
order by total_revenue desc limit 10
```

### New vs Repeat Customers


```sql
with subquery as (
  select 
    c.customer_id, 
    c.first_name,
    count(distinct f.sales_date) as visits
  from learning_pradeep.pipeline.silver_fact_sales f
  left join learning_pradeep.pipeline.silver_dim_customers c 
    on f.customer_sk = c.customer_sk
  group by c.customer_id, c.first_name
)

select 
  case 
    when visits = 1 then 'new'
    when visits > 1 and visits <= 3 then 'returning'
    else 'royal' 
  end as customer_type,
  first_name,
  count(*) as total_customers
from subquery
group by customer_type, first_name 
order by total_customers desc
limit 10
```

### Customer Spend by Location


```sql
from learning_pradeep.pipeline.silver_fact_sales f
left join learning_pradeep.pipeline.silver_dim_customers c on f.customer_sk = c.customer_sk
select c.residential_location , round(sum(f.total_amount),1) as spent

group by c.residential_location
order by spent desc limit 10
```

### Segment Performance


```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_customers c on f.customer_sk = c.customer_sk

select c.customer_segment , round(sum(f.total_amount),1) as spent

group by c.customer_segment
order by spent desc limit 10
```
