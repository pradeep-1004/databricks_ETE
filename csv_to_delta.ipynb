```python
%sql
SELECT customer_id, COUNT(*) 
FROM learning_pradeep.pipeline.bronze_dim_customers
GROUP BY customer_id
HAVING COUNT(*) > 1;
```

```python
%sql
SELECT MIN(full_date), MAX(full_date), COUNT(DISTINCT full_date)
FROM learning_pradeep.pipeline.bronze_dim_dates;
```

```python
%sql
SELECT * FROM learning_pradeep.pipeline.bronze_fact_sales_normalized LIMIT 5;
```

```python
%sql
SELECT 
SUM(CASE WHEN customer_sk IS NULL THEN 1 END) AS null_customers,
SUM(CASE WHEN product_sk IS NULL THEN 1 END) AS null_products,
SUM(CASE WHEN store_sk IS NULL THEN 1 END) AS null_stores
FROM learning_pradeep.pipeline.bronze_fact_sales_normalized;
```
