# TABLE VERIFICATION

```python
%sql
SELECT table_catalog, table_schema, table_name
FROM system.information_schema.tables
WHERE table_schema = 'pipeline';
```

# COUNT VERIFICATION NOT 0

```python
%sql
SELECT 'fact_sales' as table_name, COUNT(*) FROM learning_pradeep.pipeline.bronze_fact_sales_normalized
UNION ALL
SELECT 'dim_campaigns', COUNT(*) FROM learning_pradeep.pipeline.bronze_dim_campaigns
UNION ALL
SELECT 'dim_customers', COUNT(*) FROM learning_pradeep.pipeline.bronze_dim_customers
UNION ALL
SELECT 'dim_dates', COUNT(*) FROM learning_pradeep.pipeline.bronze_dim_dates
UNION ALL
SELECT 'dim_products', COUNT(*) FROM learning_pradeep.pipeline.bronze_dim_products
UNION ALL
SELECT 'dim_salespersons', COUNT(*) FROM learning_pradeep.pipeline.bronze_dim_salespersons
UNION ALL
SELECT 'dim_stores', COUNT(*) FROM learning_pradeep.pipeline.bronze_dim_stores;
```

# BASIC NULL CHECK 

```python
%sql
SELECT 
SUM(CASE WHEN customer_sk IS NULL THEN 1 END) AS null_customers,
SUM(CASE WHEN product_sk IS NULL THEN 1 END) AS null_products,
SUM(CASE WHEN store_sk IS NULL THEN 1 END) AS null_stores
FROM learning_pradeep.pipeline.bronze_fact_sales_normalized;
```
