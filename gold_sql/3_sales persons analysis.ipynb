### Top Salespersons

```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_salespersons s on f.salesperson_sk = s.salesperson_sk

select s.salesperson_name , round(sum(f.total_amount),1) as total_sales
group by s.salesperson_name
order by total_sales desc limit 10
```

### Role Performance

```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_salespersons s on f.salesperson_sk = s.salesperson_sk

select s.salesperson_role , round(sum(f.total_amount),1) as revenue

group by s.salesperson_role
order by revenue desc limit 10
```

Contribution %

```sql
SELECT salesperson_name,
ROUND(SUM(total_amount),1) as revenue,
SUM(total_amount) / SUM(SUM(total_amount))OVER()*100 AS contribution_percentage

FROM learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_salespersons s on f.salesperson_sk = s.salesperson_sk

GROUP BY salesperson_name
ORDER BY revenue DESC LIMIT 10
```
