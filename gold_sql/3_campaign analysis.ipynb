### Revenue vs Budget

```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_campaigns c on f.campaign_sk = c.campaign_sk

select c.campaign_name ,c.campaign_budget, round(sum(f.total_amount),1) as total_revenue

group by c.campaign_name,c.campaign_budget
order by total_revenue desc limit 10
```

### ROI

```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_campaigns c on f.campaign_sk = c.campaign_sk

select c.campaign_name , sum(f.total_amount) as total_revenue , sum(c.campaign_budget) as campaign_budget,
sum(f.total_amount) - sum(c.campaign_budget) / campaign_budget as roi

group by c.campaign_name,c.campaign_budget
order by total_revenue desc limit 10
```

### Best & Worst Campaigns

```sql
from learning_pradeep.pipeline.silver_fact_sales f 
left join learning_pradeep.pipeline.silver_dim_campaigns c on f.campaign_sk = c.campaign_sk

select c.campaign_name , round(sum(f.total_amount),1) as total_revenue 

group by c.campaign_name
--best 10
order by total_revenue desc limit 10
```
